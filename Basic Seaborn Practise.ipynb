```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Dist plot is show the distribution data as a histogram
# KDE= Karnel Density Estmation plot shows the density curve of data
# if it is set to Ture

sns.distplot(tips['total_bill'],kde=False, bins=40)
```

```python
# Jointplot shows the data of two parameter given in the argument

sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
# Pairplot vizualize the data of each pair of data column from the dataset
sns.pairplot(tips,hue='sex')
```

```python
# rugplot pint every data of given arguament like a histogram but it is not actually the histogram
sns.rugplot(tips['total_bill'])
```

```python
# Normal KDE Plot, KDE=Karnel Density Estimation plot
sns.distplot(tips['total_bill']) 
```

```python
# Categorical Plots
# I am plotting the sex categorical data
# It requires a category column and a numeric column
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
tips.tail()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
